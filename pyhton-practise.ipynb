In shallow copy, 
* in 1D array/ 1D list, it will not show any changes.
* in 2D array, nested list, any changes in oe list will reflect in "copied" list as they are sharing the same refernce, not the same values. Only references are copied, or addresses are copied of the original list. 
* If you append any values in any list, the those changes or new values appended will not reflect in other list.

In deep copy,
* it copies the values in those refernces, hence they are not sharing the same reference, hence any changes made in one list will not affect the other list.


```python
lst1=[[1,2,3,4],[5,6,7,8]]
lst3 = copy.deepcopy(lst1)
lst1, lst3
```

```python
lst1[1][0]=1000
lst1, lst3
```

```python
lst3[1][0]= 2000
lst1,lst3
```

## Decorators

A decorator in Python is a special function that can modify the behavior of another function. Think of it as a wrapper that you put around a function to add some extra functionality before and after the original function runs.


Why Use Decorators?
Decorators are useful for adding functionality to existing functions in a clean and readable way. They are often used for:

Logging
Access control
Caching
Measuring execution time

```python
def decorator(fun):
    def wrapper():
        print("happened before ")
        fun()
        print("happened after.. hahaha")
    return wrapper
        
        
def say_hello():
    print("Helloooooo")

#applying the decorator method 1
say_hello = decorator(say_hello)

#applying the decorator method 2
@decorator
def say_hello():
    print("Helloooooo")

say_hello()
```

## class methods and static methods

```python
class student:
    dept_no = "132"
    dept = "CSE"
    @classmethod
    def show_dept(self):
        return "DEPT is "+self.dept

s = student()
s.show_dept()
```

```python
class student:
    dept_name = "CSE"
    @staticmethod
    def static_method():
        return "Dep name is "+student.dept_name
    
s =student()
s.static_method()
```

## list slicing

list_name[statr_index, stop_index+1, step] 

```python
lst = [11,12,13,14,15]
slice_lst = lst[1:4]
print(slice_lst)

slice_lst2 = lst[1:4:2]
print(slice_lst2)

slice_lst3 = lst[3:]
print(slice_lst3)
```

The slicing operation lst[-2:] is used to extract a portion of the list starting from the second-to-last element to the end of the list.

Slicing: The slice lst[-2:] means “start from the element at index -2 and go to the end of the list.”

```python
lst = [11,12,13,14,15]
slice_lst = lst[-2:]
slice_lst
```
